```python
import requests
import json

url = "https://air-quality-api.open-meteo.com/v1/air-quality?latitude=19.0728&longitude=72.8826&hourly=pm10,pm2_5,nitrogen_dioxide,carbon_dioxide,dust,methane"
response = requests.get(url)

data = response.json()
#print(data)
# Extract metadata (single values)
latitude = data["latitude"]
longitude = data["longitude"]
generationtime_ms = data["generationtime_ms"]
utc_offset_seconds = data["utc_offset_seconds"]
timezone = data["timezone"]
timezone_abbreviation = data["timezone_abbreviation"]
elevation = data["elevation"]


# Extract hourly arrays
hourly = data["hourly"]
times = hourly["time"]
pm10 = hourly["pm10"]
pm25 = hourly["pm2_5"]
nitrogen_dioxide = hourly["nitrogen_dioxide"]
carbon_dioxide = hourly["carbon_dioxide"]
dust = hourly["dust"]
methane = hourly["methane"]

# Build full rows including metadata
rows = [
    (latitude, longitude, generationtime_ms, utc_offset_seconds, timezone, timezone_abbreviation, elevation, times[i], pm10[i], pm25[i],nitrogen_dioxide[i], carbon_dioxide[i], dust[i], methane[i])
    for i in range(len(times))
]

# Create Spark DataFrame
columns =  ["latitude", "longitude", "generationtime_ms", "utc_offset_seconds", "timezone", "timezone_abbreviation", "elevation", "time", "pm10", "pm2_5", "nitrogen_dioxide", "carbon_dioxide", "dust", "methane"]
df = spark.createDataFrame(rows, columns)
df.write.format("delta").mode("append").saveAsTable("air_quality_analytics_data.tbl_mumbai_air_quality")
#'latitude': 19.099998, 'longitude': 72.90001, 'generationtime_ms': 0.20432472229003906, 'utc_offset_seconds': 0, 'timezone': 'GMT', 'timezone_abbreviation': 'GMT', 'elevation': 8.0
```
